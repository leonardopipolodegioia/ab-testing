```python
import pandas as pd
from ab_test import ABTest
import plotly.graph_objects as go
import plotly.io as pio

# Needs this to make Plotly render LaTeX in the subtitles
pio.renderers.default = "notebook_connected"

# Imports the mock data
data = pd.read_csv("../data/ab_test_data.csv")

# Extracts the conversion metrics for A and B groups
a_metrics = data[data['group'] == 'A']["converted"].values
b_metrics = data[data['group'] == 'B']["converted"].values

# Sets up A/B testing
ab_test_conversion = ABTest(a_metrics, b_metrics)

# Computes and shows the observed delta
delta_obs = ab_test_conversion.compute_delta_obs()
print(f"\nObserved difference in conversion rate (B - A): {delta_obs * 100:.2f}%")

# Computes p-value on one-sided test
p_value_conversion = ab_test_conversion.compute_p_value(alternative="greater", N=1000)
print(f"P-value (right tail): {p_value_conversion:.4f}")

# 5. Distribution plot
fig_conversion = ab_test_conversion.plot_delta_distribution(N=1000, include_clt=True, alternative="greater")
fig_conversion.show()
```

```python
# Now we do it for revenues
a_metrics = data[data['group'] == 'A']['revenue'].values
b_metrics = data[data['group'] == 'B']['revenue'].values

ab_test_revenue = ABTest(a_samples=a_metrics, b_samples=b_metrics)

# Observed delta
delta_obs_revenue = ab_test_revenue.compute_delta_obs()
print(f"Mean revenue observed difference (B - A): ${delta_obs_revenue:.2f}")

# Two sided p-value
p_value_revenue = ab_test_revenue.compute_p_value(alternative="two-sided", N=1000)
print(f"P-value (two-sided): {p_value_revenue:.4f}")

if p_value_revenue < 0.05:
    print("Conclusion: Revenue difference is statistically significant (p < 0.05).")
else:
    print("Conclusion: There is not enough evidence to confirm a significant difference in revenue")

# 5. Plot
fig_revenue = ab_test_revenue.plot_delta_distribution(N=5000, include_clt=True, alternative="two-sided")
fig_revenue.show()

# 6. Exports the plot. If desired, uncomment
# fig_revenue.write_image(
#     "../img/ab_distribution.png", 
#     width=900,    
#     height=600,    
#     scale=3,        # Improve resolution
#     engine="kaleido"
# )
```
